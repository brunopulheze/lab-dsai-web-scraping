#### 2.1. List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
# your code here
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")
response
````

````python
import pandas as pd
import re

language_names = []
related_articles = []
article_counts_int = []

# Find all divs with class 'central-featured-lang'
for lang_block in soup.find_all('div', class_='central-featured-lang'):
    # Language name is in <strong>
    lang_name_tag = lang_block.find('strong')
    lang_name = lang_name_tag.get_text(strip=True) if lang_name_tag else ''

    # Article count is in <bdi>
    article_count_tag = lang_block.find('small')
    article_count = article_count_tag.get_text(strip=True) if article_count_tag else ''

    # Extract just the number from the article count string (e.g., "6,791,000+")
    num = re.sub(r'[^\d]', '', article_count)  # Remove non-digit characters
    num_int = int(num) if num else 0

    language_names.append(lang_name)
    related_articles.append(article_count)
    article_counts_int.append(num_int)

# Create DataFrame
df = pd.DataFrame({
    'Language_Name': language_names,
    'Related_Article': related_articles,
    'Article_Count_Int': article_counts_int
})

# Sort by article count descending
df_sorted = df.sort_values(by='Article_Count_Int', ascending=False).reset_index(drop=True)

print(df_sorted[['Language_Name', 'Related_Article']])
````

#### 2.2. Display the top 10 languages by number of native speakers stored in a pandas dataframe.
Hint: After finding the correct table you want to analyse, you can use a nested **for** loop to find the elements row by row (check out the 'td' and 'tr' tags). <br>An easier way to do it is using pd.read_html(), check out documentation [here](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.read_html.html).

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
# your code here
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")
response
````

````python
language_names = []
native_speakers = []

table = soup.find('table', class_='wikitable')
for row in table.find_all('tr')[1:]:  # Skip header
    cols = row.find_all('td')
    if len(cols) > 1:
        lang_name = cols[0].get_text(strip=True)
        native_speakers_count = cols[1].get_text(strip=True)
        # Clean count
        native_speakers_count = native_speakers_count.split('[')[0].replace(',', '').strip()
        language_names.append(lang_name)
        native_speakers.append(int(native_speakers_count) if native_speakers_count.isdigit() else None)

df = pd.DataFrame({
    'Language_Name': language_names,
    'Native_Speakers': native_speakers
})

df_sorted = df.sort_values(by='Native_Speakers', ascending=False).reset_index(drop=True)
print(df_sorted.head(10))
````

#### 3.1. Find the image source link and the TV show link. After you're able to retrieve, add them to your initial dataframe

````python
img_srcs = []
links = []

# Find all product cards (inspect the HTML for the correct class name)
for item in soup.find_all('div', class_='c-finderProductCard'):
    # Find the link
    a_tag = item.find('a', href=True)
    link = a_tag['href'] if a_tag else None
    links.append(link)
    
    # Find the image
    img_tag = item.find('img', class_='c-finderProductCard_mustImage')
    img_src = img_tag['src'] if img_tag else None
    img_srcs.append(img_src)

df = pd.DataFrame({
    'Link': links,
    'Image_Src': img_srcs
})
print(df)
````

#### Find the book name, price and stock availability from books to scrape website as a pandas dataframe.

````python
# This is the url you will scrape in this exercise. 
# It is a fictional bookstore created to be scraped. 
url = 'http://books.toscrape.com/'
````

````python
# your code here
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")
response
````

````python
book_names = []
book_prices = []
book_stock = []

book = soup.find_all('li')
for item in book:
    name = item.find('h3')
    price = item.find('p', class_='price_color')
    stock = item.find('p', class_='instock availability')
    
    if name and price and stock:
        book_names.append(name.get_text(strip=True))
        book_prices.append(price.get_text(strip=True))
        book_stock.append(stock.get_text(strip=True))

df = pd.DataFrame({
    'Book_Name': book_names,
    'Book_Price': book_prices,
    'Book_Stock': book_stock
})
print(df)
````

####  Display the initial 100 books available in the homepage. Once again, collect the book name, price and its stock availability.

***Hint:*** The total number of displayed books per page is 20, but you can easily move to the next page by looping through the desired number of pages and adding it to the end of the url.

````python
# This is the url you will scrape in this exercise
url = 'https://books.toscrape.com/catalogue/page-'
# This is how you will loop through each page:
number_of_pages = int(100/20)
each_page_urls = []
for n in range(1, number_of_pages+1):
    link = url+str(n)+".html"
    each_page_urls.append(link)
    
each_page_urls
````

````python
# your code here
book_names = []
book_prices = []
book_stock = []
for page_url in each_page_urls:
    response = requests.get(page_url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    
    books = soup.find_all('li')
    for item in books:
        name = item.find('h3')
        price = item.find('p', class_='price_color')
        stock = item.find('p', class_='instock availability')
        
        if name and price and stock:
            book_names.append(name.get_text(strip=True))
            book_prices.append(price.get_text(strip=True))
            book_stock.append(stock.get_text(strip=True))
df = pd.DataFrame({
    'Book_Name': book_names,
    'Book_Price': book_prices,
    'Book_Stock': book_stock
})
print(df)

````
